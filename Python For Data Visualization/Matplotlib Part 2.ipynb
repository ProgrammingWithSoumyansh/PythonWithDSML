```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,5,0.1)
y1 = x**3
y2 = x**2
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x = np.arange(0,5,0.1)
y1 = x**3
y2 = x**2
plt.plot(x,y1,color = 'black')
plt.plot(x,y2, color = 'red')
plt.show()
```

```python
x = np.arange(0,5,0.1)
y1 = x**3
y2 = x**2
plt.plot(x,y1,color = 'black', linewidth = 5)
plt.plot(x,y2, color = 'red', linewidth = 3)
plt.show()
```

```python
x = np.arange(0,5,0.1)
y1 = x**3
y2 = x**2
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y1,color = 'black', linewidth = 5, label = "x^3")
plt.plot(x,y2, color = 'red', linewidth = 3, label = "x^2")
plt.legend()
plt.show()
```

```python
x = np.arange(0,5,0.1)
y1 = x**3
y2 = x**2
plt.xlabel("x")
plt.ylabel("y")
plt.title("Square Vs Cube")
plt.plot(x,y1,color = 'black', linewidth = 5, label = "x^3")
plt.plot(x,y2, color = 'red', linewidth = 3, label = "x^2")
plt.legend()
plt.show()
```

```python
x = np.arange(0,5,0.1)
y1 = x**3
y2 = x**2
plt.xlabel("x")
plt.ylabel("y")
plt.title("Square Vs Cube")
plt.text(2,80,"Sample text", fontsize = 12)
plt.plot(x,y1,color = 'black', linewidth = 5, label = "x^3")
plt.plot(x,y2, color = 'red', linewidth = 3, label = "x^2")
plt.legend()
plt.show()
```

## Pie Graph

```python
sizes = [3,4,6,2]
```

```python
plt.pie(sizes)
plt.show()
```

```python
labels = ["A","B","C","D"]
colors = ["blue","purple","red","pink"]
plt.title("Splits among classes")
plt.pie(sizes, labels = labels, colors = colors)
plt.show()
```

```python
## Autopct -- % display
labels = ["A","B","C","D"]
colors = ["blue","purple","red","pink"]
plt.title("Splits among classes")
plt.pie(sizes, labels = labels, colors = colors, autopct = "%.1f%%")
plt.show()
```

```python
## explode
```

```python
## Autopct -- % display
labels = ["A","B","C","D"]
colors = ["blue","purple","red","pink"]
explode = [0.1,0,0,0]
plt.title("Splits among classes")
plt.pie(sizes, labels = labels, colors = colors, autopct = "%.1f%%", explode = explode)
plt.show()
```

```python
## Autopct -- % display
labels = ["A","B","C","D"]
colors = ["blue","purple","red","pink"]
explode = [0.1,0.2,0,0]
plt.title("Splits among classes")
plt.pie(sizes, labels = labels, colors = colors, autopct = "%.1f%%", explode = explode)
plt.show()
```
