## Single plot display of multiple coordinates

```python
x1 = [1,2,3]
y1 = [2,4,6]

x2 = [4,5,6]
y2 = [8,10,12]

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

## Customisation

```python
x1 = [1,2,3]
y1 = [2,4,6]

x2 = [4,5,6]
y2 = [8,10,12]

plt.plot(x1,y1,'r')
plt.plot(x2,y2,'g')
plt.show()
```

```python
x1 = [1,2,3]
y1 = [2,4,6]

x2 = [4,5,6]
y2 = [8,10,12]

plt.plot(x1,y1,'r+')
plt.plot(x2,y2,'g-.')
plt.show()
```

```python
x = np.array([1,2,3,4])
y = x**3
plt.plot(x,y)
plt.show()
```
