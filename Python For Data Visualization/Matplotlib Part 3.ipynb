## Histograms

```python
## Frequency distribution --- no of times a certain element/info has occured
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 x = [1, 2, 3, 1, 2, 3, 4, 1, 3, 4, 5]
 plt.hist(x, orientation="vertical")
 plt.show()
```

## Bargraphs

```python
year = ['year 2017','year 2018','year 2019','year 2020']
sal = [100,200,300,400]
plt.bar(year, sal, width = 0.3)
plt.xticks(rotation = 40)
plt.show()
```

```python
## When to use?
Line Scatter Bar Plot -- X and Y data 
Scatter -- for correlation between two different values
Pie Chart -- % distribution -- Govt Spending -- Defense, Railway, Education 
```
